```python
#nullData > Funkcija koja prebrojava koliko ima null vrednosti meu obeleijim
def nullData(data):
    missing = data.isna().sum()
    print(missing)
```

```python
#editDta > funkcija koja vri male modifikacije nad skupom podataka radi lake analize
def editDta(data):
    datae = data
    datae.loc[datae['PURCHASES_FREQUENCY']==0.0, 'PURCHASES_FREQUENCY'] = 0.0
    datae.loc[datae['PURCHASES_FREQUENCY']>0.0, 'PURCHASES_FREQUENCY'] = 1.0

    datae.loc[datae['ONEOFF_PURCHASES_FREQUENCY']==0.0, 'ONEOFF_PURCHASES_FREQUENCY'] = 0.0
    datae.loc[datae['ONEOFF_PURCHASES_FREQUENCY']>0.0, 'ONEOFF_PURCHASES_FREQUENCY'] = 1.0
    
    return datae
```

Kako te null vrednosti mogu pozitivno ili negativno uticati na proces klasteriyaije, moemo primeniti jednu od tri tehnika za "otklanjanje" null vrednosti.

### Tehnike za otklanjanje null vrednosti:
1. Prvi nain za otklanjanje vrednosti koje nedostaju jeste ta da obriemo ceo uzorak koji ima null vrednist.

2. Druga tehnika za reavanje ovog problema je kreiranje odvojenog modela.

3. Trea tehnika se zasniva na statistikim metodama.

#### Brisanje uzorka

Najjednostavni nain za reavanje ovakvog problema je, kako smo rekli, brisanje svih uzoraka koji poseduju null vrednost. Prednost ovog pristupa je jednostavnost, dok je mana to to se bez velikih posledica na tanost modela, ovaj pristup moe upotrebiti na velike skupove podataka sa oko minimum 1000000 uzoraka.


#### Kreiranje odvojenog modela

Kreiranjem zasebnog modela moemo doi do vrednosti koje nam nedostaju i to na sledei nain:

1. Prvo odvajamo uzorke iz skupa koji nemaju null vredosti u skup pdataka za trening.

2. Potom odvojimo obeleije novog trening skupa koje sadri null vrednost od ostalih, jer emo to obeleije koristiti kao ciljnu vrednost u zasebnom modelu.

3. Kreiramo zasebni model i treniramo na novom trenig skupu podataka.

4. Nakon treniranja modela, koristrimo odvojene podatke iz skupa koji sadre null vrednosti, proputamo ih kroz model kako bi smo dobili izlaz sa kojim emo popuniti null vrednost.

#### Nedostaci tehnike kreiranja odvojenog modela

Nedostaci ovog metoda jssu, utroak vremena na treniranje odvojenog modela i dobijanje vrednosti.
Ako skup podataka ima vie obeleija kod kojih postoje null vrednosti, morali bi za svako obeleije kreirati i trenirati odvojeni model, tako da nije preporujivo koristiti kod velikih skupova podataka.

#### Statistike metode

Ovo je najee koriena metoda. Kod ove metode se null vrednosti obeleija popunjavaju:

1. Aritmetikom sredinom vrednosti obeleija(mean).

2. Sredinom vrednosti obeleija. Sortiraju se vrednosti potom se bira vrednost u sredini(median).

3. Vrednost sa najveom frekvencijom(mode).

Kako skup podataka koji analiziramo nije veliki, neemo brisati uzorke koji sadre null vrednosti. Takoe, imamo dva obeleija koja sadre null vrednosti:

to znai da bi smo morali kreirati dva odvojena modela, trenirati ih i potom skup podataka punili vrednostima, to je malo sloeniji zadatak. Zbog svega navedenog emo za nau analizu koristiti statistiu metodu aritmetike sredine.

# K-means algoritam

K-means algoritam je algoritam nenadgledanog uenja koji na osnovu obeleija podataka grupie podatke u k grupa. Ovaj algoritam se izvrava u nekoliko sledeih koraka:

1. Ulazni parametar jeste broj k(broj grupa) koji korisnik sam zadaje.

2. Nasumice ili po doreenom algoritmu biranje centroida za svaku od k grupa.

3. Odreivanje broja iteracija algoritma.

4. U svakoj od iteracija algoritam izvrava korakle:
    
      1. Za svaki uzorak iz skupa uzoraka, rauna se, po odreenoj metrici slinost, odnosno                rastojanje izmeu uzorka i centara svih k grupa.
      
      2. Uzorak e se, u trenutnoj iteraciji, pridruiti grupi sa ijim centroidom ima najveu              slinost.
      


Pre nego to primenimo kmeans klasterizaciju, potruduemo se da pronaemo broj klastera tj. K koji bi bio najoptimalni, ali prvo da skaliramo vrednosti naeg skupa podataka.

```python
#ScaleData > Funkcija koja skalira podatke na opseg vrednosti [-1 1]
def ScaleData(data):
    sc = StandardScaler()
    scaledData = sc.fit_transform(data)
    return scaledData
```

```python
scaledData = ScaleData(data)
datas = pd.DataFrame(scaledData, columns=data.columns)
datas
```

Pogledajmo korelaciju izmeu obeleija.

```python
#printCorelation > Funkcija koja tampa korelaciju izmeu obeleija
def printCorelation(data):
    plt.figure(figsize = (12, 12))
    sn.heatmap(data.corr(), annot=True, cmap='viridis', 
    xticklabels=data.columns,
    yticklabels=data.columns)
    
printCorelation(datas)    
```

Za izbor najoptimalnog broja klastera koristiemo najpoznatiji elbow method.

```python
#FindOptimalK > Funkcija koja pronalazi noptimalan broj klastera
def FindK(data):
    scores = []
    rangeValues = range(1, 10)
    for i in rangeValues:
        Kmeans = KMeans(n_clusters = i)
        Kmeans.fit(data)
        scores.append(Kmeans.inertia_)
    return scores
    
#DrawGraphicK > Funkcija koja iscrtava grafik izbora broja klastera
def DrawGraphicK(SArray, RangeNumbers):
    plt.plot(RangeNumbers, SArray)
    plt.xlabel('Broj klastera k')
    plt.ylabel('Poeni')
    plt.show()
```

```python
rv = range(1, 10)
#Scores = FindK(scaledData)
Scores = FindK(scaledData)
DrawGraphicK(Scores, rv)
```

moramo da izaberemo vrednost k na laktu, tj. taku nakon koje izoblienje / inercija poinju da se linearno smanjuju. U naem sluaju e to biti broj 4.

Kod Kmeans algoritma je dosta bitno inicijalni izbor centroida klastera, tako da emo za iniciranje centroida koristiti kmeans++ algoritam.

```python
Kmeans = KMeans(n_clusters = 4, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 101)
predict = Kmeans.fit_predict(scaledData)
print(np.unique(predict))
data["cluster"] = predict
sn.scatterplot(data = data, x= 'BALANCE', y='PAYMENTS', hue= "cluster")
```

Analiza glavnih komponenti (eng. principal component analysis, skraeno PCA) je statistika procedura kojom se skup od eventualno korelisanih atributa, preslikava u skup nekorelisanih atributa. Svaki od atributa iz novog skupa se naziva glavna komponenta, a svi oni su linearno nekorelisani. Prva glavna komponenta se bira tako da ima najveu moguu varijansu (drugim reima, bira se tako da du nje podaci najvie variraju). Svaka naredna glavna komponenta se bira tako da je normalna na sve prethodne i da, takoe, ima trenutnu najveu varijansu. Na ovaj nain se smanjuje dimenzionalnost poetnog prostora. Pritom se deo informacija moe izgubiti.

```python
#selectPCA > Funkcija koja selektuje obeleija uz pomo pca algoritma
def selectPCA(data, n):
    pca = PCA(n_components = n)
    principal_comp = pca.fit_transform(data)
    lab = pca.components_
    r = range(n)
    pca_df = pd.DataFrame(data = principal_comp, columns = [*r])
    pca_df = pd.concat([pca_df, pd.DataFrame({'cluster': predict})], axis = 1)
    return lab, pca_df
```

```python
lab, pca = selectPCA(data, 5)
```

```python
best_cols = [0,1,2, 3, 4]
Kmeans = KMeans(n_clusters = 4, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 101)
best_vals = pca[best_cols].iloc[ :, 1:].values
predict = Kmeans.fit_predict(best_vals)

pca["cluster"] = predict
best_cols.append("cluster")
sn.pairplot(pca[best_cols], hue="cluster")
```

Hajde sada da samo inicijativno selektujemo odreena obeleija i videti kako e kmeans izvriti klasterizaciju.

```python


best_cols = ["BALANCE", "PURCHASES", "CASH_ADVANCE","CREDIT_LIMIT", "PAYMENTS", "MINIMUM_PAYMENTS"]
kmeans = KMeans(n_clusters=4, init="k-means++", n_init=4, max_iter=300) 
best_vals = data[best_cols].iloc[ :, 1:].values
y_pred = kmeans.fit_predict(best_vals)

data["cluster"] = y_pred
best_cols.append("cluster")
sn.pairplot(data[best_cols], hue="cluster")

```

# Hierarhisko klasteroanje
Jedna od najpriridnih metoda klasteriyacije jeste upravo hierarhiska metoda. Kod hierarhiskog klasterovanja imamo dva pristupa, a to su:
1. Top-down
2. Bottom-up



#### Top-down pristup
Kod Top-down pristupa uzima se ceo skup podataka kao poetni klaster i u svakom narednom koraku se rekurzivno skup podataka deli na manje podskupove. Ova vrsta klasterovanja se naziva i klasterizacija deljenjem.

#### Bottom-up
Kod bottom-up pristupa kree se od svakog elementa skupa podataka kao klastera i u svakom narednom koraku se srodni klasteri spajaju u jedan i pomeraju hierarhiju gore. Ovakvo klasterovanje podataka se jo naziva Aglomerativno klasterovanje.

```python
hc = AgglomerativeClustering(n_clusters = 3, affinity = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(scaledData)
```

```python


#Visualizing the clusters 
best_cols = ["BALANCE", "PURCHASES", "CASH_ADVANCE","CREDIT_LIMIT", "PAYMENTS", "MINIMUM_PAYMENTS"]
hc = AgglomerativeClustering(n_clusters = 4, affinity = 'euclidean', linkage = 'ward')
best_vals = data[best_cols].iloc[ :, 1:].values
y_hc = hc.fit_predict( best_vals )

data["cluster"] = y_hc
best_cols.append("cluster")
sn.pairplot( data[ best_cols ], hue="cluster")

```
