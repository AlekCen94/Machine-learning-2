```python
#nullData > Funkcija koja prebrojava koliko ima null vrednosti među obeležijim
def nullData(data):
    missing = data.isna().sum()
    print(missing)
```

```python
#editDta > funkcija koja vrši male modifikacije nad skupom podataka radi lakše analize
def editDta(data):
    datae = data
    datae.loc[datae['PURCHASES_FREQUENCY']==0.0, 'PURCHASES_FREQUENCY'] = 0.0
    datae.loc[datae['PURCHASES_FREQUENCY']>0.0, 'PURCHASES_FREQUENCY'] = 1.0

    datae.loc[datae['ONEOFF_PURCHASES_FREQUENCY']==0.0, 'ONEOFF_PURCHASES_FREQUENCY'] = 0.0
    datae.loc[datae['ONEOFF_PURCHASES_FREQUENCY']>0.0, 'ONEOFF_PURCHASES_FREQUENCY'] = 1.0
    
    return datae
```

Kako te null vrednosti mogu pozitivno ili negativno uticati na proces klasteriyaije, možemo primeniti jednu od tri tehnika za "otklanjanje" null vrednosti.

### Tehnike za otklanjanje null vrednosti:
1. Prvi način za otklanjanje vrednosti koje nedostaju jeste ta da obrišemo ceo uzorak koji ima null vrednist.

2. Druga tehnika za rešavanje ovog problema je kreiranje odvojenog modela.

3. Treća tehnika se zasniva na statističkim metodama.

#### Brisanje uzorka

Najjednostavni način za rešavanje ovakvog problema je, kako smo rekli, brisanje svih uzoraka koji poseduju null vrednost. Prednost ovog pristupa je jednostavnost, dok je mana to što se bez velikih posledica na tačnost modela, ovaj pristup može upotrebiti na velike skupove podataka sa oko minimum 1000000 uzoraka.


#### Kreiranje odvojenog modela

Kreiranjem zasebnog modela možemo doći do vrednosti koje nam nedostaju i to na sledeći način:

1. Prvo odvajamo uzorke iz skupa koji nemaju null vredosti u skup pdataka za trening.

2. Potom odvojimo obeležije novog trening skupa koje sadrži null vrednost od ostalih, jer ćemo to obeležije koristiti kao ciljnu vrednost u zasebnom modelu.

3. Kreiramo zasebni model i treniramo na novom trenig skupu podataka.

4. Nakon treniranja modela, koristrimo odvojene podatke iz skupa koji sadrže null vrednosti, propuštamo ih kroz model kako bi smo dobili izlaz sa kojim ćemo popuniti null vrednost.

#### Nedostaci tehnike kreiranja odvojenog modela

Nedostaci ovog metoda jssu, utrošak vremena na treniranje odvojenog modela i dobijanje vrednosti.
Ako skup podataka ima više obeležija kod kojih postoje null vrednosti, morali bi za svako obeležije kreirati i trenirati odvojeni model, tako da nije preporuččjivo koristiti kod velikih skupova podataka.

#### Statističke metode

Ovo je najčešće korišćena metoda. Kod ove metode se null vrednosti obeležija popunjavaju:

1. Aritmetičkom sredinom vrednosti obeležija(mean).

2. Sredinom vrednosti obeležija. Sortiraju se vrednosti potom se bira vrednost u sredini(median).

3. Vrednost sa najvećom frekvencijom(mode).

Kako skup podataka koji analiziramo nije veliki, nećemo brisati uzorke koji sadrže null vrednosti. Takođe, imamo dva obeležija koja sadrže null vrednosti:
1. CREDIT_LIMIT(Ograničenje kreditne kartice korisnika).

2. MINIMUM_PAYMENTS(Minimalni iznos uplata koje je izvršio korisnik).

Što znači da bi smo morali kreirati dva odvojena modela, trenirati ih i potom skup podataka punili vrednostima, što je malo složeniji zadatak. Zbog svega navedenog ćemo za našu analizu koristiti statističu metodu aritmetičke sredine.

# Kasterizacija
U uvodu smo reki da problem klasterizacije spada u domen nenadgledanog mašinskog učenja. Šta je zapravo klasterizacija? Za klasterizaciju bi smo mogli reći da ne predstavlja ništa drugo do grupisanja objekata. 
#### Gupisanje


Zadatak grupisanja bio bi da postavi objekat u grupu gde se naalaze objekti slični njemu.

# K-means algoritam

K-means algoritam je algoritam nenadgledanog učenja koji na osnovu obeležija podataka grupiše podatke u k grupa. Ovaj algoritam se izvršava u nekoliko sledećih koraka:

1. Ulazni parametar jeste broj k(broj grupa) koji korisnik sam zadaje.

2. Nasumice ili po doređenom algoritmu biranje centroida za svaku od k grupa.

3. Određivanje broja iteracija algoritma.

4. U svakoj od iteracija algoritam izvršava korakle:
    
      1. Za svaki uzorak iz skupa uzoraka, računa se, po određenoj metrici sličnost, odnosno                rastojanje između uzorka i centara svih k grupa.
      
      2. Uzorak će se, u trenutnoj iteraciji, pridružiti grupi sa čijim centroidom ima najveću              sličnost.
      


Pre nego što primenimo kmeans klasterizaciju, potrudućemo se da pronađemo broj klastera tj. K koji bi bio najoptimalni, ali prvo da skaliramo vrednosti našeg skupa podataka.

```python
#ScaleData > Funkcija koja skalira podatke na opseg vrednosti [-1 1]
def ScaleData(data):
    sc = StandardScaler()
    scaledData = sc.fit_transform(data)
    return scaledData
```

```python
scaledData = ScaleData(data)
datas = pd.DataFrame(scaledData, columns=data.columns)
datas
```

Pogledajmo korelaciju između obeležija.

```python
#printCorelation > Funkcija koja štampa korelaciju između obeležija
def printCorelation(data):
    plt.figure(figsize = (12, 12))
    sn.heatmap(data.corr(), annot=True, cmap='viridis', 
    xticklabels=data.columns,
    yticklabels=data.columns)
    
printCorelation(datas)    
```

Za izbor najoptimalnog broja klastera koristićemo najpoznatiji elbow method.

```python
#FindOptimalK > Funkcija koja pronalazi noptimalan broj klastera
def FindK(data):
    scores = []
    rangeValues = range(1, 10)
    for i in rangeValues:
        Kmeans = KMeans(n_clusters = i)
        Kmeans.fit(data)
        scores.append(Kmeans.inertia_)
    return scores
    
#DrawGraphicK > Funkcija koja iscrtava grafik izbora broja klastera
def DrawGraphicK(SArray, RangeNumbers):
    plt.plot(RangeNumbers, SArray)
    plt.xlabel('Broj klastera k')
    plt.ylabel('Poeni')
    plt.show()
```

```python
rv = range(1, 10)
#Scores = FindK(scaledData)
Scores = FindK(scaledData)
DrawGraphicK(Scores, rv)
```

moramo da izaberemo vrednost k na „laktu“, tj. tačku nakon koje izobličenje / inercija počinju da se linearno smanjuju. U našem slučaju će to biti broj 4.

Kod Kmeans algoritma je dosta bitno inicijalni izbor centroida klastera, tako da ćemo za iniciranje centroida koristiti kmeans++ algoritam.

```python
Kmeans = KMeans(n_clusters = 4, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 101)
predict = Kmeans.fit_predict(scaledData)
print(np.unique(predict))
data["cluster"] = predict
sn.scatterplot(data = data, x= 'BALANCE', y='PAYMENTS', hue= "cluster")
```

Analiza glavnih komponenti (eng. principal component analysis, skraćeno PCA) je statistička procedura kojom se skup od eventualno korelisanih atributa, preslikava u skup nekorelisanih atributa. Svaki od atributa iz novog skupa se naziva glavna komponenta, a svi oni su linearno nekorelisani. Prva glavna komponenta se bira tako da ima najveću moguću varijansu (drugim rečima, bira se tako da duž nje podaci najviše variraju). Svaka naredna glavna komponenta se bira tako da je normalna na sve prethodne i da, takođe, ima trenutnu najveću varijansu. Na ovaj način se smanjuje dimenzionalnost početnog prostora. Pritom se deo informacija može izgubiti.

```python
#selectPCA > Funkcija koja selektuje obeležija uz pomoć pca algoritma
def selectPCA(data, n):
    pca = PCA(n_components = n)
    principal_comp = pca.fit_transform(data)
    lab = pca.components_
    r = range(n)
    pca_df = pd.DataFrame(data = principal_comp, columns = [*r])
    pca_df = pd.concat([pca_df, pd.DataFrame({'cluster': predict})], axis = 1)
    return lab, pca_df
```

```python
lab, pca = selectPCA(data, 5)
```

```python
best_cols = [0,1,2, 3, 4]
Kmeans = KMeans(n_clusters = 4, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 101)
best_vals = pca[best_cols].iloc[ :, 1:].values
predict = Kmeans.fit_predict(best_vals)

pca["cluster"] = predict
best_cols.append("cluster")
sn.pairplot(pca[best_cols], hue="cluster")
```

Hajde sada da samo inicijativno selektujemo određena obeležija i videti kako će kmeans izvršiti klasterizaciju.

```python


best_cols = ["BALANCE", "PURCHASES", "CASH_ADVANCE","CREDIT_LIMIT", "PAYMENTS", "MINIMUM_PAYMENTS"]
kmeans = KMeans(n_clusters=4, init="k-means++", n_init=4, max_iter=300) 
best_vals = data[best_cols].iloc[ :, 1:].values
y_pred = kmeans.fit_predict(best_vals)

data["cluster"] = y_pred
best_cols.append("cluster")
sn.pairplot(data[best_cols], hue="cluster")

```

# Hierarhisko klasteroanje
Jedna od najpriridnih metoda klasteriyacije jeste upravo hierarhiska metoda. Kod hierarhiskog klasterovanja imamo dva pristupa, a to su:
1. Top-down
2. Bottom-up



#### Top-down pristup
Kod Top-down pristupa uzima se ceo skup podataka kao početni klaster i u svakom narednom koraku se rekurzivno skup podataka deli na manje podskupove. Ova vrsta klasterovanja se naziva i klasterizacija deljenjem.

#### Bottom-up
Kod bottom-up pristupa kreće se od svakog elementa skupa podataka kao klastera i u svakom narednom koraku se srodni klasteri spajaju u jedan i pomeraju hierarhiju gore. Ovakvo klasterovanje podataka se još naziva Aglomerativno klasterovanje.

```python
hc = AgglomerativeClustering(n_clusters = 3, affinity = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(scaledData)
```

```python


#Visualizing the clusters 
best_cols = ["BALANCE", "PURCHASES", "CASH_ADVANCE","CREDIT_LIMIT", "PAYMENTS", "MINIMUM_PAYMENTS"]
hc = AgglomerativeClustering(n_clusters = 4, affinity = 'euclidean', linkage = 'ward')
best_vals = data[best_cols].iloc[ :, 1:].values
y_hc = hc.fit_predict( best_vals )

data["cluster"] = y_hc
best_cols.append("cluster")
sn.pairplot( data[ best_cols ], hue="cluster")

```
